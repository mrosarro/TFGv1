```python
import pandas as pd
```

```python
weather = pd.read_csv('weather.csv', dtype={'column7': float, 'column8': float})
flights = pd.read_csv('flights_transformed.csv')
```

```python
print(flights)
```

```python
print(weather)
```

```python
flights.info()
```

```python
flights['DATE']=flights['DATE'].astype('datetime64')
flights.info()
```

```python
weather['date']=weather['date'].astype('datetime64')
weather.info()
```

```python
print(flights)
```

```python
print(weather)
```

```python
dftotal2015=flights.merge(weather, left_on=['ORIGIN_AIRPORT', 'DATE', 'TIME2'], right_on=['airport', 'date', 'time'], how='left')
```

```python
print(dftotal2015)
```

```python
result = dftotal2015[dftotal2015['ORIGIN_AIRPORT'] == 'ORD']
result_sorted = result.sort_values(by='TIME2')
pd.set_option('display.max_columns', None)
print(result_sorted)
```

```python
dftotal2015.to_csv('data_combined.csv', index=False)
```
